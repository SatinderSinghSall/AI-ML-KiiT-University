## Experiment - 1:
- Write a program to reverse a given integer using a while loop.
- Example: Input: 456, Output: 654

```python
# Program to reverse a number using a while loop

num = int(input("Enter a number: "))

reverse = 0

while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num //= 10

print("Reversed number is:", reverse)
```

## Experiment - 2:
- Write a program to count the number of digits in a given number using a while loop.
- Example: Input: 78945, Output: 5

```python
# Program to count the number of digits using a while loop

num = int(input("Enter a number: "))

count = 0

while num > 0:
    count += 1
    num //= 10

print("Number of digits:", count)
```

## Experiment - 3:
- Write a program to print the multiplication table of a given number using a while loop.
- Example: Input: 7. Output: 7 × 1 = 7 … 7 × 10 = 70

```python
# Program to print multiplication table using a while loop

num = int(input("Enter a number: "))

i = 1
while i <= 10:
    print(num, "x", i, "=", num * i)
    i += 1
```

## Experiment - 4:
- Write a program to find the sum of the digits of a given number using a while loop.
- Example: Input: 1234, Output: 10

```python
# Program to find the sum of digits using a while loop

num = int(input("Enter a number: "))

sum_digits = 0

while num > 0:
    digit = num % 10
    sum_digits += digit
    num //= 10

print("Sum of digits is:", sum_digits)
```

## Experiment - 5:
- Write a program to calculate the factorial of a given number using a while loop.
- Example: Input: 5, Output: 120

```python
# Program to calculate factorial using a while loop

num = int(input("Enter a number: "))

factorial = 1
i = num

while i > 0:
    factorial *= i
    i -= 1

print("Factorial of", num, "is", factorial)
```
