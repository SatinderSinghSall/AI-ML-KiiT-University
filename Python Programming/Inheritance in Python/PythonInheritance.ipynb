```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)


x = Person("John", "Doe")
x.printname()
```

```python
class Vechale:
    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed
    
    def show(self):
        print(f"Brand: {self.brand}")
        print(f"Speed: {self.speed}")

class Car(Vechale):
    def __init__(self, brand, speed, fuel_type,):
        super().__init__(brand, speed)
        self.fuel_type = fuel_type
    
    def show(self):
        super().show()
        print(f"Fuel Type: {self.fuel_type}")


car = Car("Toyota", 120, "Petrol")
car.show()
```

```python
class Employee:
    def __init__(self, emp_id, salary):
        self.emp_id = emp_id
        self.salary = salary

class Manager(Employee):
    def __init__(self, emp_id, salary, department):
        super().__init__(emp_id, salary)
        self.department = department

    def displayData(self):
        print(f"Employee ID: {self.emp_id}")
        print(f"Salary: {self.salary}")
        print(f"Department: {self.department}")

manager = Manager(101, "50000", "Software Development")
manager.displayData()
```

```python
class Shape:
    def area(self):
        print("Area is not defined.")

class Rectangle(Shape):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth

    def area(self):
        print(f"Area of Rectangle: {self.length * self.breadth}")

r = Rectangle(5, 3)
r.area()
```

```python
class Account:
    def __init__(self, account_no, balance):
        self.account_no = account_no
        self.balance = balance

    def display_balance(self):
        print(f"Account Number: {self.account_no}")
        print(f"Balance: {self.balance}")

class SavingsAccount(Account):
    def add_interest(self):
        interest = self.balance * 0.06
        self.balance = self.balance + interest
        print(f"Balance after 6% interest: {self.balance}")

s = SavingsAccount(12345, 10000)
s.display_balance()
s.add_interest()
```

```python
class Mobile:
    def __init__(self, brand, price):
        self.brand = brand
        self.price = price

class SmartPhone(Mobile):
    def __init__(self, brand, price, camara_megapixels):
        super().__init__(brand, price)
        self.camara_megapixels = camara_megapixels

    def displayData(self):
        print(f"Mobile Brand: {self.brand}")
        print(f"Price: {self.price}")
        print(f"Camara Megapixels: {self.camara_megapixels}")

phone = SmartPhone("Samsung", 90000, 50)
phone.displayData()
```
