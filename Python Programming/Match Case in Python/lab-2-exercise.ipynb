## Exercise - 1: Simple calculator using match-case:

```python
num1 = float(input("Enter the number 1: "))
num2 = float(input("Enter the number 1: "))
opt = input("Enter operator: ")

match opt:
    case '+':
        result = num1 + num2;
    case '-':
        result = num1 - num2;
    case '*':
        result = num1 * num2;
    case '/':
        result = num1 / num2;
    case _:
        printf("Invalid Entry, try again.")

print(result)
```

## Exercise - 2: Vowel and Consonant:

```python
ch = input("Enter a word: ").lower()

match ch:
    case 'a' | 'e' | 'i' | 'o' | 'u':
        print("Vowel")
    case _:
        print("Consonant OR Invalid Input.")
```

## Exercise - 3: Grading System:

```python
grade = input("Enter your grade (A / B / C / D / F) : ").upper()

match grade:
    case 'A':
        print("Excellent. As always.")
    case 'B':
        print("Very Good. Keep it up.")
    case 'C':
        print("Good. Can do better.")
    case 'D':
        print("Pass. Need more improvement.")
    case 'F':
        print("Fail! Try again.")
    case _:
        print("Invalid Entry!")
```

## Exercise - 4: Traffic Light System:

```python
traffic_light = input("Enter traffic light (R / G / Y) : ").upper()

match traffic_light:
    case 'R':
        print("RED LIGHT...STOP")
    case 'G':
        print("Green Light...GO")
    case 'Y':
        print("Yellow Light...GET READY")
    case _:
        print("Invalid Entry!")
```

## Exercise - 5: Number Classification:

```python
num = float(input("Enter a number: "))

match num:
    case x if x > 0:
        print("Positive Number.")
    case x if x < 0:
        print("Negetive Number.")
    case x if x == 0:
        print("Zero.")
    case _:
        print("Invalid Entry.")
```

## Exercise - 7: Basic ATM Operations:

```python
balance = 5000  # initial balance
1

print("ATM Menu:")
print("1. Balance Inquiry")
print("2. Cash Withdrawal")
print("3. Deposit")

choice = int(input("Enter your choice (1-3): "))

match choice:
    case 1:
        print("Your Balance =", balance)
    case 2:
        amount = float(input("Enter amount to withdraw: "))
        if amount <= balance and amount > 0:
            balance -= amount
            print("Withdrawal successful!")
            print("Remaining Balance =", balance)
        else:
            print("Insufficient balance or invalid amount")
    case 3:
        amount = float(input("Enter amount to deposit: "))
        if amount > 0:
            balance += amount
            print("Deposit successful!")
            print("Updated Balance =", balance)
        else:
            print("Invalid deposit amount")
    case _:
        print("Invalid choice")
```
