## Lab Experiment - 2:
- Q2. Write a Python program to find the length of a list entered by the user.

```python
list2 = input("Enter list item: ")
my_list = list2.split()

print(my_list)
print(f"The list length: {len(my_list)}")
```

## Lab Experiment - 3:
- Q3. Write a Python program to access and display the first and last elements of a list.

```python
list3 = ["Satinder", "Nick", "Charlie", "Vaibhav"]

print(f"The list first element is: {list3[0:1]}")
print(f"The list last element is: {list3[3:]}")
```
