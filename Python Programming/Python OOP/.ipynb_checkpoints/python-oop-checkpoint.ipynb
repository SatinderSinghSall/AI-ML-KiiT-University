```python
class Book:
    def __init__(self, book_title, author, price):
        self.book_title = book_title
        self.author = author
        self.price = price

    def display_book(self):
        print("Title:", self.book_title)
        print("Author:", self.author)
        print("Price:", self.price)

    def discount_price(self, percent):
        discount = (percent/100) * self.price
        final_price = self.price - discount
        print("Original Price:", self.price)
        print("Discount:", percent, "%")
        print("Final Price:", final_price)


# Example
b1 = Book("Python", "John", 500)
b1.display_book()
b1.discount_price(10)
```

```python
class Result:
    def __init__(self, student_name, math_marks, physics_marks, computer_marks):
        self.student_name = student_name
        self.math_marks = math_marks
        self.physics_marks = physics_marks
        self.computer_marks = computer_marks

    def total_marks(self):
        return self.math_marks + self.physics_marks + self.computer_marks

    def percentage(self):
        return self.total_marks() / 3

    def result(self):
        if self.math_marks >= 40 and self.physics_marks >= 40 and self.computer_marks >= 40:
            return "Pass"
        else:
            return "Fail"


# Example
r1 = Result("Rahul", 75, 60, 80)
print("Total:", r1.total_marks())
print("Percentage:", r1.percentage())
print("Result:", r1.result())
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = {}

    def add_item(self, item_name, price):
        self.items[item_name] = price

    def remove_item(self, item_name):
        if item_name in self.items:
            del self.items[item_name]

    def calculate_total(self):
        total = sum(self.items.values())
        print("Total Price:", total)


# Example
cart = ShoppingCart()
cart.add_item("Shirt", 800)
cart.add_item("Shoes", 1500)
cart.remove_item("Shirt")
cart.calculate_total()
```

```python
import math

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius * self.radius

    def circumference(self):
        return 2 * math.pi * self.radius


# Example
c1 = Circle(7)
print("Area:", c1.area())
print("Circumference:", c1.circumference())
```

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius

    def to_fahrenheit(self):
        return (self.celsius * 9/5) + 32

    def to_kelvin(self):
        return self.celsius + 273.15


# Example
t1 = Temperature(25)
print("Fahrenheit:", t1.to_fahrenheit())
print("Kelvin:", t1.to_kelvin())
```
